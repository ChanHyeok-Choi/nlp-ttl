## 2. 문장에 대한 확률
* 'An adorable little boy is spreading smiles'의 확률: P(An adorable little boy is spreading smiles)
    
    

* 문장의 확률을 구하기 위해서 각 단어에 대한 예측 확률들을 곱합니다.

## 3. 카운트 기반의 접근
* An adorable little boy가 나왔을 때, is가 나올 확률: P(is|An adorable little boy)

    

* 기계가 학습한 말뭉치 데이터에서 An adorable little boy가 100번 등장했을 때 is가 등장한 경우가 30번이라면, 이 경우 위의 확률은 30%입니다.

## 4. 카운트 기반 접근의 한계 - 희소 문제: Sparsity Problem
* 언어 모델은 실생활에서 사용되는 언어의 확률 분포를 근사 모델링 합니다. 
* 기계에게 많은 코퍼스를 훈련시켜서 언어 모델을 통해 현실에서의 확률 분포를 근사하는 것이 언어 모델의 목표입니다.
* 그렇다면 기계가 훈련하는 데이터는 정말 방대한 양이 필요해집니다.
* 예를 들어, 위의 예시 확률을 구하는 경우에서 훈련 데이터에 An adorable little boy is 라는 단어 시퀀스가 없다면 확률은 0이 됩니다. 또는 An adorable little boy라는 단어 시퀀스가 없다면 분모가 0이 되어 확률이 정의되지 않습니다.
* 그렇다면 해당 단어 시퀀스가 ㅇ벗다고 해서 이 확률을 0 혹은 정의되지 않는 확률이라고 하는 것이 정확한 모델링 방법일까요?
* NO! 현실에선 해당 단어 시퀀스가 존재하기에 정답일 가능성이 높습니다.
* 위와 같이 **충분한 데이터를 관측하지 못하여 언어를 정확히 모델링하지 못하는 문제를 희소성 문제** 라고 합니다.

## 5. 해결 방법은?
* 위 문제를 완화하는 방법으로 n-gram 언어 모델이나 스무딩, 백오프와 같은 여러가지 일반화(normalization) 기법이 존재합니다.
* 하지만 희소성 문제에 대한 근본적인 해결책이 되진 못했습니다.
* 결국 이러한 한계에 부딪혀 인공 신경망 언어 모델로 트렌드가 바뀌게 됩니다.
