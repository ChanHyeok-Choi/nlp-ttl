# Perplexity, PPL
* 일일히 모델들에 대해 실제 작업을 시켜보고 정확도를 비교하는 것은 공수가 너무 많이 드는 작업입니다. 또한, 모델 수가 많아지면 시간도 그 만큼 늘어나게 됩니다.
* 그래서 이러한 방식보단 조금 부정확할 수 있어도 테스트 데이터에 대해서 빠르게 식으로 계산되는 간단한 평가 방법이 바로 PPL 입니다.
* 즉, 바로 모델 내에서 자신의 성능을 수치화해서 결과를 내놓습니다.

## 2. 분기 계수: Branching factor
* PPL은 또한 선택할 수 있는 경우의 수를 의미하는 분기계수(branching factor)입니다. 즉, 특정 시점에서 평균적으로 몇 개의 선택지를 가지고 고민하고 있는지를 의미합니다.
* 예를 들어, PPL=10이 나왔다면 해당 언어 모델은 다음 단어를 예측하는 모든 시점마다 평균 10개의 단어를 가지고 어떤 것인지 정답인지 고민하고 있다고 볼 수 있습니다. 그래서 PPL이 낮은 모델이 더 성능이 좋다고 볼 수 있습니다:

    
* 하지만 조심해야 할 점은 PPL이 낮은 것은 테스트 데이터 상 높은 정확도를 보인다는 것이지 사람이 직접 느끼기에 좋은 언어 모델이 아니라는 것입니다.
* 또한, PPL은 테스트 데이터에 의존하므로 두 개 이상의 언어 모델을 비교할 땐 정량적으로 양이 많고, 도메인에 알맞은 동일한 테스트 데이터를 사용해야 신뢰도가 높다는 것입니다.
