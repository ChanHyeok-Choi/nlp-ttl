# 한국어에서의 언어 모델: LM for Korean Sentences
* 영어보다 한국어가 다음 단어를 예측하기가 훨씬 까다롭습니다.

## 1. 한국어는 어순이 중요하지 않다.
* 그래서 이전 단어가 주어졌을 때 다음 단어가 나타날 확률을 구해야 하는데 어순이 중요하지 않다는 것은 다음 단어로 어떤 단어든 등장할 수 있다는 의미입니다.
* 예시:
    1. 나는 운동을 합니다 체육관에서.  
    2. 나는 체육관에서 운동을 합니다.  
    3. 체육관에서 운동을 합니다.  
    4. 나는 운동을 체육관에서 합니다.
* 4개의 문장은 전부 의미가 통하고, 심지어 '나는'을 생략해도 말이 되어버립니다. 이런 특성 때문에 다음 단어를 예측하기가 어렵습니다.

## 2. 한국어는 교착어이다.
* 한국어에는 조사가 있지만 영어는 기본적으로 조사가 없습니다. 그래서 한국어에는 어떤 행동을 하는 동사의 주어나 목적어를 위해 조사라는 것이 있습니다.
* 예를 들어, '그녀'만 해도 '그녀가, 그녀의, 그녀와, 그녀로, 그녀께서, 그녀처럼' 등과 같이 다양한 경우가 존재합니다.
* 그렇기 때문에 한국어에선 **토큰화**를 통해 접사나 조사 등을 분리하는 작업이 중요합니다.

## 3. 한국어는 띄어쓰기가 제대로 지켜지지 않는다.
* 한국어는 띄어쓰기를 제대로 하지 않아도 의미가 전달되며, 띄어쓰기 규칙 또한 상대적으로 까다로운 언어입니다.
* 그래서 NLP에 있어서 한국어 코퍼스는 띄어쓰기가 제대로 지켜지지 않는 경우가 많습니다.
* 토큰이 제대로 분리되지 않은채 훈련 데이터로 사용된다면 언어 모델은 제대로 작동하지 않습니다.
