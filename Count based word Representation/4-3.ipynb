## 1. DTM의 표기법
* 문서 단어 행렬(DTM)이란 다수의 문서에서 등장하는 각 단어들의 빈도를 행렬로 표현한 것
* 쉽게 생각하면 각 문서에 대한 BoW를 하나의 행렬로 만든 것!

## 2. DTM의 한계
1. 희소 표현: Sparse representation

    원-핫 벡터와 마찬가지로 전체 코퍼스가 방대한 데이터라면 문서 벡터의 차원은 수만 이상의 차원을 가질 수도 있습니다. 또한 많은 문서 벡터가 대부분의 값이 0일 수도 있습니다. 이런 것들을 희소 벡터(sparse vector) 또는 희소 행렬(sparse matrix)라고 부르는데, 많은 양의 저장 공간과 높은 계산 복잡도를 요구합니다.
2. 단순 빈도 수 기반 접근:
    
    예를 들어 영어에 대해서 DTM을 만들었을 때, 불용어인 the는 어떤 문서이든 자주 등장할 수 밖에 없습니다. 그런데 유사한 문서인지 비교하고 싶은 문서1, 문서2, 문서3에서 동일하게 the가 빈도수가 높다고 해서 이 문서들이 유사한 문서라고 판단해서는 안 됩니다. 그래서 불용어와 중요한 단어에 대해서 가중치를 줄 수 있는 방법이 TF-IDF입니다.
