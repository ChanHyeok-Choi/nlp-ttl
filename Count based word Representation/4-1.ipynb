# 다양한 단어의 표현 방법

## 1. 단어의 표현 방법
* 국소 표현(Local Representation) or 이산 표현(Discrete Representation) 방법: 해당 단어 그 자체만 보고, 특정 값을 매핑하여 단어를 표현
    * 예시: puppy(강아지), cute(귀여운), lovely(사랑스러운)라는 단어가 있을 때 각 단어에 1번, 2번, 3번 등과 같은 숫자를 맵핑(mapping)하여 부여
* 분산 표현(Distributed Representation) or 연속 표현(Continuous Representation) 방법: 그 단어를 표현하고자 주변을 참고하여 단어를 표현
    * puppy(강아지)라는 단어 근처에는 주로 cute(귀여운), lovely(사랑스러운)이라는 단어가 자주 등장하므로, puppy라는 단어는 cute, lovely한 느낌이다로 단어를 정의합니다.
    * 국소 표현 방법은 단어의 의미, 뉘앙스를 표현할 수 없는 반면 분산 표현 방법은 단어의 뉘앙스를 표현할 수 있습니다.

## 2. 단어 표현의 카테고리화
* Bag of Words는 국소 표현에 속하며, 단어의 빈도수를 카운트하여 단어를 수치화하는 단어 표현 방법입니다.
* 이 BoW의 확장 버전이 Document-Term Matrix, DTM 이고, 이러한 빈도수 기반 단어 표현에 단어의 중요도에 따른 가중치를 줄 수 있는 것이 TF-IDF 입니다.
* 워드 임베딩은 연속 표현에 속하며 예측을 기반으로 단어의 뉘앙스를 표현하는 Word2Vec과 그의 확장인 FastText가 있고, 예측과 카운트 모두를 사용하는 GloVe가 있습니다.
