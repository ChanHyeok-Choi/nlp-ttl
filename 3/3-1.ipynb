## 1. 단어 시퀀스의 할당
* 기계 번역: Machine Translation, P(나는 버스를 탔다) > P(나는 버스를 태운다)
* 오타 교정: Spell Correction, 선생님이 교실로 부리나케 P(달려갔다) > P(잘려갔다)
* 음성 인식: Speech Recognition, P(나는 메롱을 먹는다) < P(나는 메론을 먹는다)
