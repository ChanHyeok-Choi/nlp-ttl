## 2. 시그모이드 함수: Sigmoid function
* 수식:

    

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x, y, 'g')
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(0.5*x)
y2 = sigmoid(x)
y3 = sigmoid(2*x)

plt.plot(x, y1, 'r', linestyle='--') # w의 값이 0.5일때 빨간선
plt.plot(x, y2, 'g') # w의 값이 1일때 초록선
plt.plot(x, y3, 'b', linestyle='--') # w의 값이 2일때 파란선
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y1 = sigmoid(x+0.5)
y2 = sigmoid(x+1)
y3 = sigmoid(x+1.5)

plt.plot(x, y1, 'r', linestyle='--') # x + 0.5 빨간선
plt.plot(x, y2, 'g') # x + 1 초록선
plt.plot(x, y3, 'b', linestyle='--') # x + 1.5 파란선
plt.plot([0,0],[1.0,0.0], ':') # 가운데 점선 추가
plt.title('Sigmoid Function')
plt.show()
```

## 3. 비용 함수: Cost function
* 평균 제곱 오차를 로지스틱 회귀의 비용 함수로 사용했을 때 좋지 않은 로컬 미니멈에 빠질 가능성이 높아 문제 해결이 어렵습니다.

    
* J는 목적 함수를 의미하고 f가 실제값 y와 예측값 H(x)의 오차를 나타내는 함수라고 할 때, f를 어떻게 정의하느냐에 따라 적절한 목적 함수가 완성됩니다.

    
    
    
* y의 실제값이 1일 때 -logH(x)의 그래프(파란선)를 사용하고, 0일 때 -log(1-H(x)) 그래프(빨간선)를 사용해야 합니다.

    
    
* 결국 로지스틱 회귀의 목적 함수는 다음과 같습니다.
