# 정규표현식: Regular Expression

## 1. 정규표현식 문법
![Alt text](image.png)


```python
import re
```

```python
# .기호
r = re.compile("a.c")
r.search("kkk") # 아무런 결과도 출력되지 않는다.
r.search("abc")
```

```python
# ?기호
r = re.compile("ab?c")
r.search("abbc") # 아무런 결과도 출력되지 않는다.
r.search("abc") # <re.Match object; span=(0, 3), match='abc'>
r.search("ac")
```

```python
# *기호
r = re.compile("ab*c")
r.search("a") # 아무런 결과도 출력되지 않는다.
r.search("ac") # <re.Match object; span=(0, 2), match='ac'>
r.search("abc") # <re.Match object; span=(0, 3), match='abc'>
r.search("abbbbc")
```

```python
# +기호
r = re.compile("ab+c")
r.search("ac") # 아무런 결과도 출력되지 않는다.
r.search("abc") # <re.Match object; span=(0, 3), match='abc'>
r.search("abbbbc")
```

```python
# ^기호
r = re.compile("^ab")

# 아무런 결과도 출력되지 않는다.
r.search("bbc")
r.search("zab")

r.search("abz")
```

```python
# {숫자} 기호
r = re.compile("ab{2}c")

# 아무런 결과도 출력되지 않는다.
r.search("ac")
r.search("abc")
r.search("abbbbbc")

r.search("abbc")

# {숫자1, 숫자2} 기호
r = re.compile("ab{2,8}c")

# 아무런 결과도 출력되지 않는다.
r.search("ac")
r.search("abc")
r.search("abbbbbbbbbc")

r.search("abbc") # <re.Match object; span=(0, 4), match='abbc'>

r.search("abbbbbbbbc")
```

```python
# {숫자,} 기호
r = re.compile("a{2,}bc")

# 아무런 결과도 출력되지 않는다.
r.search("bc")
r.search("aa")

r.search("aabc") # <re.Match object; span=(0, 4), match='aabc'>

r.search("aaaaaaaabc")
```

```python
# []기호
r = re.compile("[abc]") # [abc]는 [a-c]와 같다.
r.search("zzz") # 아무런 결과도 출력되지 않는다.

# <re.Match object; span=(0, 1), match='a'>
r.search("a")
r.search("aaaaaaa")

r.search("baac")      
```

```python
# [^문자]기호
r = re.compile("[^abc]")

# 아무런 결과도 출력되지 않는다.
r.search("a")
r.search("ab") 
r.search("b")

r.search("d") # <re.Match object; span=(0, 1), match='d'>

r.search("1")                                                                                                
```

## 2. 정규표현식 모듈 함수
![Alt text](image-1.png)

```python
# re.match()와 re.search()의 차이
r = re.compile("ab.")
r.match("kkkabc") # 아무런 결과도 출력되지 않는다.

r.search("kkkabc") # <re.Match object; span=(3, 6), match='abc'>

r.match("abckkk")
```

```python
# re.split()

# 공백 기준 분리
text = "사과 딸기 수박 메론 바나나"
re.split(" ", text)

# 줄바꿈 기준 분리
text = """사과
딸기
수박
메론
바나나"""
re.split("\n", text)

# '+'를 기준으로 분리
text = "사과+딸기+수박+메론+바나나"
re.split("\+", text)
```

```python
# re.findall()
text = """이름 : 김철수
전화번호 : 010 - 1234 - 1234
나이 : 30
성별 : 남"""

re.findall("\d+", text)
```

```python
re.findall("\d+", "문자열입니다.")
```

```python
# re.sub()
text = "Regular expression : A regular expression, regex or regexp[1] (sometimes called a rational expression)[2][3] is, in theoretical computer science and formal language theory, a sequence of characters that define a search pattern."

preprocessed_text = re.sub('[^a-zA-Z]', ' ', text)
print(preprocessed_text)
print([word for word in preprocessed_text.split(' ') if word != ''])
```

### 정규 표현식의 전처리 예제

```python
text = """100 John    PROF
101 James   STUD
102 Mac   STUD"""
re.split('\s+', text)
```

```python
re.findall('\d+', text)
```

```python
re.findall('[A-Z]',text)
```

```python
re.findall('[A-Z]{4}',text)
```

```python
re.findall('[A-Z][a-z]+',text)
```

### 정규 표현식을 이용한 토큰화

```python
# NLTK에서는 정규 표현식을 사용해서 단어 토큰화를 수행하는 RegexpTokenizer를 지원합니다.
from nltk.tokenize import RegexpTokenizer

text = "Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop"

tokenizer1 = RegexpTokenizer("[\w]+")
tokenizer2 = RegexpTokenizer("\s+", gaps=True)

print(tokenizer1.tokenize(text))
print(tokenizer2.tokenize(text))
```
